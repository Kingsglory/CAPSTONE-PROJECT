```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\emeka\OneDrive\Desktop\DATA GRAPH 2.csv")

df
```

```python
plt.figure(figsize=(15,8))

sns.lineplot(data=df.iloc[:, :-1])

plt.ylabel('MURDER')

plt.title('ASCENDING DATAFRAM PLOT OF DISTRICT')
plt.xticks(rotation=90)
```

```python
sns.catplot(data=df,x='STATE',y='MURDER', kind='bar')

plt.xticks(rotation=90, fontsize=(13))

plt.figure(figsize=(18,30))

plt.tight_layout()
```
